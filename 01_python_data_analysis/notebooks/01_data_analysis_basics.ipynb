# Python 数据分析入门

本教程将带你学习使用 Pandas 进行数据分析的基础操作。

## 1. 导入必要的库

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## 2. 加载数据

```python
df = pd.read_csv('../data/sales_data.csv')
df.head()
```

## 3. 数据基本信息

```python
print('数据形状:', df.shape)
print('\n数据类型:')
print(df.dtypes)
print('\n基本统计信息:')
print(df.describe())
```

## 4. 数据处理 - 添加销售额列

```python
df['total_sales'] = df['quantity'] * df['price']
df.head()
```

## 5. 按产品类别分析销售额

```python
category_sales = df.groupby('category')['total_sales'].sum().sort_values(ascending=False)
print('按类别销售总额:')
print(category_sales)
```

## 6. 可视化 - 类别销售额对比

```python
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar')
plt.title('各类别销售额对比')
plt.xlabel('产品类别')
plt.ylabel('销售额 (元)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 7. 按地区分析

```python
region_sales = df.groupby('region')['total_sales'].sum().sort_values(ascending=False)
print('按地区销售总额:')
print(region_sales)
```

## 8. 保存处理后的数据

```python
df.to_csv('../data/sales_data_processed.csv', index=False)
print('数据已保存!')
```
