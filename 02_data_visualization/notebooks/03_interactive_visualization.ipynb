# 交互式数据可视化

本教程将学习使用Plotly创建交互式数据可视化。

## 1. 导入必要的库

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## 2. 加载数据

```python
# 加载天气数据
df = pd.read_csv('../data/weather_data.csv')
df['date'] = pd.to_datetime(df['date'])
print('数据形状:', df.shape)
print('数据预览:')
print(df.head())
```

## 3. 交互式折线图

```python
# 创建交互式折线图
fig = px.line(df, x='date', y='temperature', color='city', title='各城市温度变化趋势', labels={'date': '日期', 'temperature': '温度 (°C)', 'city': '城市'}, hover_data=['humidity', 'wind_speed'])

# 美化图表
fig.update_layout(xaxis_title='日期', yaxis_title='温度 (°C)', legend_title='城市', template='plotly_white', hovermode='x unified')

# 显示图表
fig.show()
```

## 4. 交互式散点图

```python
# 创建交互式散点图
fig = px.scatter(df, x='temperature', y='humidity', color='city', size='wind_speed', hover_name='date', title='温度与湿度关系', labels={'temperature': '温度 (°C)', 'humidity': '湿度 (%)', 'wind_speed': '风速 (km/h)'})

fig.update_layout(xaxis_title='温度 (°C)', yaxis_title='湿度 (%)', legend_title='城市', template='plotly_white')

fig.show()
```

## 5. 交互式柱状图

```python
# 按城市和日期计算平均温度
city_date_avg = df.groupby(['city', 'date']).mean().reset_index()

# 创建交互式柱状图
fig = px.bar(city_date_avg, x='date', y='temperature', color='city', title='各城市每日平均温度', labels={'date': '日期', 'temperature': '平均温度 (°C)', 'city': '城市'})

fig.update_layout(xaxis_title='日期', yaxis_title='平均温度 (°C)', legend_title='城市', template='plotly_white')

fig.show()
```

## 6. 交互式热力图

```python
# 准备热力图数据
heatmap_data = df.pivot(index='city', columns='date', values='temperature')

# 创建交互式热力图
fig = px.imshow(heatmap_data, title='各城市温度热力图', labels={'x': '日期', 'y': '城市', 'color': '温度 (°C)'}, color_continuous_scale='RdBu_r')

fig.update_layout(template='plotly_white')

fig.show()
```

## 7. 交互式仪表盘

```python
# 创建子图仪表盘
fig = make_subplots(rows=2, cols=2, subplot_titles=('温度变化', '湿度分布', '风速变化', '降水量分布'))

# 添加温度折线图
for city in df['city'].unique():
    city_data = df[df['city'] == city]
    fig.add_trace(go.Scatter(x=city_data['date'], y=city_data['temperature'], name=city), row=1, col=1)

# 添加湿度箱线图
fig.add_trace(go.Box(y=df['humidity'], x=df['city'], name='湿度'), row=1, col=2)

# 添加风速折线图
for city in df['city'].unique():
    city_data = df[df['city'] == city]
    fig.add_trace(go.Scatter(x=city_data['date'], y=city_data['wind_speed'], name=city), row=2, col=1)

# 添加降水量柱状图
fig.add_trace(go.Bar(x=df['date'], y=df['precipitation'], name='降水量'), row=2, col=2)

# 更新布局
fig.update_layout(height=800, width=1000, title_text='天气数据仪表盘', template='plotly_white', hovermode='x unified')

fig.show()
```

## 8. 3D 散点图

```python
# 创建3D散点图
fig = px.scatter_3d(df, x='temperature', y='humidity', z='wind_speed', color='city', size='precipitation', hover_name='date', title='天气数据3D可视化')

fig.update_layout(scene=dict(xaxis_title='温度 (°C)', yaxis_title='湿度 (%)', zaxis_title='风速 (km/h)'), template='plotly_white')

fig.show()
```
