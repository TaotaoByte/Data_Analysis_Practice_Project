# 数据可视化入门

本教程将带你学习使用 Matplotlib 和 Seaborn 进行数据可视化。

## 1. 导入必要的库

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

## 2. 加载数据

```python
df = pd.read_csv('../data/weather_data.csv')
df.head()
```

## 3. 柱状图 - 各城市平均温度

```python
avg_temp = df.groupby('city')['temperature'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_temp.index, y=avg_temp.values, palette='coolwarm')
plt.title('各城市平均温度对比')
plt.xlabel('城市')
plt.ylabel('平均温度 (°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4. 折线图 - 温度随时间变化

```python
plt.figure(figsize=(12, 6))
for city in df['city'].unique():
    city_data = df[df['city'] == city]
    plt.plot(city_data['date'], city_data['temperature'], marker='o', label=city)

plt.title('各城市温度变化趋势')
plt.xlabel('日期')
plt.ylabel('温度 (°C)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. 散点图 - 温度与湿度的关系

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='temperature', y='humidity', hue='city', s=100, alpha=0.7)
plt.title('温度与湿度的关系')
plt.xlabel('温度 (°C)')
plt.ylabel('湿度 (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## 6. 热力图 - 各指标相关性

```python
corr_matrix = df[['temperature', 'humidity', 'wind_speed', 'precipitation']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', square=True)
plt.title('气象指标相关性热力图')
plt.tight_layout()
plt.show()
```

## 7. 箱线图 - 各城市温度分布

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='city', y='temperature', palette='Set3')
plt.title('各城市温度分布箱线图')
plt.xlabel('城市')
plt.ylabel('温度 (°C)')
plt.tight_layout()
plt.show()
```

## 8. 饼图 - 各城市降水量占比

```python
city_precipitation = df.groupby('city')['precipitation'].sum()

plt.figure(figsize=(10, 10))
plt.pie(city_precipitation.values, labels=city_precipitation.index, autopct='%1.1f%%', 
        startangle=90, colors=sns.color_palette('pastel'))
plt.title('各城市降水量占比')
plt.axis('equal')
plt.show()
```
